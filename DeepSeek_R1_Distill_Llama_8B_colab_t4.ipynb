```python
from transformers import pipeline
import torch
```

```python
pipe = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Llama-8B", torch_dtype=torch.float16, device_map="auto")
```

```python
messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe(messages)
```
